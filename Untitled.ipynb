```python
names = [
    'Alpha', 'Beta', 'Callisto', 'Draco', 'Gemini',
    'Hydra', 'Lynx', 'Orion', 'Phoenix', 'Ursa'
]
```

```python
net = tpnet.Net(len(names), names, max_random_edges=14)
```

```python
sname = 'Draco'
source = net.namelup[sname]
tname = 'Beta'
target = net.namelup[tname]
route = np.array([target])
sg = gt.search.dijkstra_iterator(
    net.g, net.vweight, source=source, 
    array=True
)
while route[0] != source:
    found = False
    for src, dest in sg:
        if dest == target:
            route = np.append(src, route)
            target = src
            found = True
    if not found:
        raise RuntimeError('cannot find route')
```
